```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("kyanyoga/sample-sales-data")

print("Path to dataset files:", path)
```

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
try:
    data = pd.read_csv("/kaggle/input/sample-sales-data/sales_data_sample.csv", encoding='utf-8')
except UnicodeDecodeError:
    data = pd.read_csv("/kaggle/input/sample-sales-data/sales_data_sample.csv", encoding='latin-1')

# 2Ô∏è Display first few rows
print("Dataset Preview:")
print(data.head())

#  Basic dataset info
print("\nDataset Info:")
print(data.info())

#  Summary statistics for numeric columns
print("\nSummary Statistics:")
print(data.describe())

#  Convert ORDERDATE to datetime
data['ORDERDATE'] = pd.to_datetime(data['ORDERDATE'], errors='coerce')

#  Extract Year and Month for analysis
data['Year'] = data['ORDERDATE'].dt.year
data['Month'] = data['ORDERDATE'].dt.month

#  Average sales by product line
avg_sales_product = data.groupby("PRODUCTLINE")["SALES"].mean().sort_values(ascending=False)
print("\nAverage Sales by Product Line:")
print(avg_sales_product)

#  Plot histogram of SALES
plt.figure(figsize=(8,5))
plt.hist(data["SALES"], bins=10, color="lightgreen", edgecolor="black")
plt.title("Distribution of Total Sales")
plt.xlabel("Total Sales Amount ($)")
plt.ylabel("Number of Orders")
plt.grid(True, alpha=0.3)
plt.show()

#  Plot histogram of QUANTITYORDERED
plt.figure(figsize=(8,5))
plt.hist(data["QUANTITYORDERED"], bins=10, color="skyblue", edgecolor="black")
plt.title("Distribution of Quantity Ordered")
plt.xlabel("Quantity Ordered")
plt.ylabel("Number of Orders")
plt.grid(True, alpha=0.3)
plt.show()

# Sales trend by month
monthly_sales = data.groupby("Month")["SALES"].sum()
plt.figure(figsize=(8,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='orange')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales ($)")
plt.grid(True, alpha=0.3)
plt.show()

#  Optional: Average sales by year
yearly_sales = data.groupby("Year")["SALES"].sum()
plt.figure(figsize=(8,5))
plt.bar(yearly_sales.index, yearly_sales.values, color='purple', edgecolor='black')
plt.title("Yearly Total Sales")
plt.xlabel("Year")
plt.ylabel("Total Sales ($)")
plt.grid(True, alpha=0.3)
plt.show()
```
